```python
import sklearn 
from skimage import color
from skimage import io
from matplotlib import pyplot as plt
import numpy as np
import os
import cv2
```

```python
img_O= cv2.imread("C:/Users/qamre/Downloads/sampleForLab.jpeg")
plt.imshow(img_O)
```

```python
img_g =color.rgb2gray(img_O)
plt.imshow(img_g)
img_c=color.gray2rgb(img_g)
plt.imshow(img_c)
```

```python
img_gray = cv2.cvtColor(img_O, cv2.COLOR_BGR2GRAY)
plt.imshow(img_gray)
```

```python
img_rgb = cv2.cvtColor(img_gray, cv2.COLOR_GRAY2RGB)
plt.imshow(img_rgb)
```

```python
b_channel,g_channel,r_channel = cv2.split(img_O)    # channels split,ND ARRAY  , reverse order mei tuple ko store aur access karta hai. toh thats why pehla blue green and red instead of rgb
cv2.imshow("Red", r_channel)
cv2.imshow("Green", g_channel)
cv2.imshow("Blue", b_channel)
cv2.merge([b_channel,g_channel,r_channel])
plt.imshow(img_O)
```

```python
b_channel, g_channel,r_channel = np.split(img_O, 3, axis=-1)
merged_image = np.concatenate([b_channel, g_channel, r_channel], axis=-1)
def show_images(images, titles):
    plt.figure(figsize=(12, 6))
    for i, (image, title) in enumerate(zip(images, titles)):
        plt.subplot(1, len(images), i+1)
        plt.imshow(image, cmap='gray')
        plt.title(title)
        plt.axis('off')
    plt.show()

show_images([r_channel.squeeze(), g_channel.squeeze(), b_channel.squeeze()], ['Red Channel', 'Green Channel', 'Blue Channel'])
plt.imshow(merged_image)
plt.title('Merged Image')
plt.axis('off')
plt.show()
```
